```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.relplot(x="sepal_length", y="sepal_width", data=df);
plt.show()
```

```python
sns.relplot(x="sepal_length", y="sepal_width", hue="species", data=df);
plt.show()
```

```python
sns.relplot(x="sepal_length", y="sepal_width", hue="species",
            size="petal_width", data=df);
plt.show()
```

```python
sns.relplot(x="sepal_length", y="sepal_width", hue="petal_length",
            size="petal_width", style="species", data=df);
plt.show()
```

```python
sns.histplot(data=df) # y="sepal_width")
```
